```python
from pandas import read_csv
import numpy as np
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
dataframe = read_csv("machine_learning/Classification/iris.csv")
dataframe.head()
dataset = dataframe.values
data = dataset[:,0:4].astype(float)
labels = dataset[:,4]
```

```python
(trainData, testData, trainLabels, testLabels) = train_test_split(np.array(data),
	np.array(labels), test_size=0.25, random_state=42)
```

```python
model = KNeighborsClassifier(n_neighbors=9,weights='distance')
model.fit(trainData, trainLabels)
print("KNN model result")
print(model.score(trainData, trainLabels))
print(model.score(testData, testLabels))
predictions = model.predict(testData)
print(classification_report(testLabels, predictions))

kernels = ("linear", "poly", "sigmoid", "rbf")
# "linear", "poly", "sigmoid", "rbf" , "precomputed")
values = []

for i in kernels:
  model = SVC(kernel=i,C=1)
  model.fit(trainData, trainLabels)
  val = f"Kernal = {i}"
  val1 = model.score(trainData, trainLabels)
  val2 = model.score(testData, testLabels)
  if val1 - val2 <=.03:
        values.append((val, val1, val2))
print("SVM model result")
print(values)
```

```python
features = len(dataframe.columns)
values = []
criterions = ("gini","entropy")
for i in range(1,features):
  for x in range(1,15):
    for y in criterions:
      model = DecisionTreeClassifier(random_state=42,criterion=y, max_features=i, max_depth = x)
      model.fit(trainData, trainLabels)
      val = f"Result {i} {x} {y}"
      val1 =(model.score(trainData, trainLabels))
      val2 = (model.score(testData, testLabels))
      if model.score(trainData, trainLabels) - model.score(testData, testLabels) <=.03:
        values.append((val, val1, val2))
print("Decision Tree model result")
print(values)
print(max(values))


# le = LabelEncoder()
# dataframe['gender'] = le.fit_transform(dataframe['gender'])
# data = dataframe.loc[:, ['age', 'gender', 'height_cm', 'weight_kg',
#        'body fat_%', 'diastolic', 'systolic', 'gripForce', 'sit and bend forward_cm',
#        'sit-ups counts', 'broad jump_cm']]
# labels = dataframe['class']

#splitter = best, random | max_features = int, auto, log, none | max_depth = int
features = len(dataframe. columns)
values = []
criterions = ("gini","entropy")
for i in range(30,100):
  for x in range(1,15):
    for y in criterions:
      model = RandomForestClassifier(n_estimators=i, random_state=42,criterion=y,max_depth=x)
      model.fit(trainData, trainLabels)
      val = f"Result {i} {x} {y}"
      val1 =(model.score(trainData, trainLabels))
      val2 =(model.score(testData, testLabels))
      predictions = model.predict(testData)
      # print(classification_report(testLabels, predictions))
      if model.score(trainData, trainLabels) >= 0.74 and model.score(testData, testLabels) >= 0.74:
        values.append((val, val1, val2))
print("Random Forest model result")
print(values)
print(max(values))
```

```python
print("Result of Logistic Regression")
#penality = l1,l2, elasticnet
solver = ["liblinear", "sag", "saga", "lbfgs", "newton-cg",]
#max-iter = 100
# for i in range(50,100):
for x in solver:
  try:
      if x == "liblinear":
        lib = ['l1', 'l2']
        for p in lib:
          model = LogisticRegression(penalty=p , max_iter=100 ,solver = x)
          model.fit(trainData, trainLabels)
          # if model.score(trainData, trainLabels) >= 0.80 and model.score(testData, testLabels) >= 0.80:
          print(model.score(trainData, trainLabels),x)
          print(model.score(testData, testLabels))
          predictions = model.predict(testData)
          print(classification_report(testLabels, predictions))
      elif x == "saga":
        saga = ['l1', 'l2']
        for p in saga:
          model = LogisticRegression(penalty=p , max_iter=100 ,solver = x)
          model.fit(trainData, trainLabels)
          # if model.score(trainData, trainLabels) >= 0.80 and model.score(testData, testLabels) >= 0.80:
          print(model.score(trainData, trainLabels),x)
          print(model.score(testData, testLabels))
          predictions = model.predict(testData)
          print(classification_report(testLabels, predictions))
      else:
          model = LogisticRegression(max_iter=100 ,solver = x)
          model.fit(trainData, trainLabels)
          # if model.score(trainData, trainLabels) >= 0.80 and model.score(testData, testLabels) >= 0.80:
          print(model.score(trainData, trainLabels),x)
          print(model.score(testData, testLabels))
          predictions = model.predict(testData)
          print(classification_report(testLabels, predictions))
  except ConvergenceWarning:
    pass
warnings.resetwarnings()
```
